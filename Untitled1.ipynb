**Data Pre-processing** - As we can see that the column course_students_enrolled is of the type object. We have to convert that.
There are 2 ways to do that-:

```python
# method 1

temp = []
for i in range(0, len(data)):
    try:
        num = data['course_students_enrolled'].str.split('k')[i][0]
        num = float(num) * 1000 
        temp.append(num)
    except:
        num = data['course_students_enrolled'].str.split('m')[i][0]
        num = float(num) * 1000000
        temp.append(num)
data['course_students_enrolled'] = temp
data['course_students_enrolled'] = data['course_students_enrolled'].astype(float)
```

```python
data['course_students_enrolled']
```

```python
#method 2
data1= pd.read_csv('/content/coursea_data.csv')
data1['course_students_enrolled']=data1['course_students_enrolled'].str.replace('k','*1000')
data1['course_students_enrolled']=data1['course_students_enrolled'].str.replace('m','*1000000')
data1['course_students_enrolled']=data1['course_students_enrolled'].map(lambda x: eval(x))
data1['course_students_enrolled']
```

**Which organization has the most number of courses?**

```python
co= data.groupby(['course_organization'])['course_organization'].count()
co=pd.DataFrame(co)
co=co.rename(columns={"course_organization": "Count"})
co=co.sort_values(by=['Count'], ascending=False)
co.head(1)
```

```python
co_top10=co.head(10)
plt.figure(figsize=(26,14))
p=sns.lineplot(data=co_top10)
p
```

**Which organization has the highest course rating?**

```python
cr= data.groupby(['course_organization'])['course_rating'].mean()
cr=pd.DataFrame(cr)
cr=cr.rename(columns={"course_rating": "Mean"})
cr=cr.sort_values(by=['Mean'], ascending=False)
cr.head()
```

```python
plt.figure(figsize=(154,126))
sns.barplot(x=cr.index, y=cr['Mean'])
```

**Top 5 courses on Coursera**
top 5 courses on Coursera by course rating

```python
ct= data.groupby(['course_title'])['course_rating'].mean()
ct=pd.DataFrame(ct)
ct=ct.rename(columns={"course_rating": "Mean"})
ct=ct.sort_values(by=['Mean'], ascending=False)
ct.head()
```

**Does course difficulty affect number of students enrolled?**
For this we will check the average number of students enrolled grouping by difficulty.

```python
cd= data.groupby(['course_difficulty'])['course_students_enrolled'].mean()
cd=pd.DataFrame(cd)
cd=cd.rename(columns={"course_students_enrolled": "Count"})
cd=cd.sort_values(by=['Count'], ascending=False)
cd.head()
```

**Does certificate type impact number of students enrolled?**

```python
cert_type= data.groupby(['course_Certificate_type'])['course_students_enrolled'].mean()
cert_type=pd.DataFrame(cert_type)
cert_type=cert_type.rename(columns={"course_students_enrolled": "Count"})
cert_type=cert_type.sort_values(by=['Count'], ascending=False)
cert_type.head()
```
